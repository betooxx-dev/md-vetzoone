**Listamos los directorios en la carpeta**

```python
import os 

directories = os.listdir()
for directory in directories:
    print(directory)
```

**Listar los archivos disponibles en la carpeta de encuestas**

```python
folder  = "encuestas"
directories = os.listdir(os.path.join(folder))
for directory in directories:
    print(directory)
```

**Seleccionamos el archivo de interés**

```python
file = directories[0]
path = os.path.join(folder, file)
```

```python
import pandas as pd
respuestas = pd.read_csv(path)
respuestas.head()
```

```python
cols = respuestas.columns
for col in cols:
    print(col)
```

```python
mapeo = [
    "time",
    "rango_empleo",
    "situaciones_habilidades",
    "prog_desarrollo",
    "importancia_habilidades",
    "habilidades_general",
    "habilidad_sin_dominio",
    "tiempo_atencion_semanal",
    "preferencias_act_aprendizaje",
    "frec_cursos_softskills",
    "motivacion_software",
    "funcionalidades_aprendizaje",
    "plataformas_usadas",
    "softvstech",
    "herramientas_eval",
    "uso_IA"
]

respuestas.columns = mapeo
cols = respuestas.columns
for col in cols:
    print(col)
```

```python
features_list = list(cols)
features_list.remove("time")
respuestas = respuestas[features_list]
n_respuestas, n_features = respuestas.shape
print(f"Hay {n_respuestas} respuestas para {n_features} preguntas sin considerar la marca temporal")
```

```python
respuestas.head()
```

```python
respuestas.info()
```

```python
cols = respuestas.columns
escalas = {}

for col in cols:
    escalas[col] = respuestas[col].unique()
    print(escalas[col])
    
```

```python
respuestas.describe()
opciones = respuestas.iloc[:,12].unique()
for opcion in opciones:
    print(opcion)
```

```python
def mapear_importancia(opcion):
    if opcion == "Mucho menos importantes que las técnicas":
        return 0
    elif opcion == "Algo menos importantes":
        return 1
    elif opcion == "Igual de importantes":
        return 2
    elif opcion == "Algo más importantes":
        return 3
    elif opcion == "Mucho más importantes":
        return 4

current_feature = features_list[12]
respuestas[f"{current_feature}_map"] = respuestas[current_feature].apply(mapear_importancia)
respuestas.head()
    
```

```python
respuestas.describe()
# si 0 = 0 y 4 = 100 cuánto es 2.281250?
media =  respuestas.iloc[:,-1].mean()
nivel = 25*media
print(nivel)
```

```python
import seaborn as sns
from matplotlib import pyplot as plt

plt.figure(figsize=(5, 3))
sns.histplot(respuestas.iloc[:,12])
resumen = f"Nivel de importancia {nivel:.2f}%"
plt.legend([resumen])
plt.xticks(rotation=45)
#plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
from matplotlib import pyplot as plt

plt.figure(figsize=(6, 6))
importancia = respuestas.groupby([current_feature]).count()
print(importancia.shape)
plt.pie(importancia.iloc[:,-1])
resumen = f"Nivel de importancia {nivel:.2f}%"
plt.legend([resumen])
plt.xticks(rotation=45)
#plt.tight_layout()
plt.show()
```
